## Ejercicio 4: Derivada Numérica por Diferencias Centradas

### Objetivo
Aproximar la derivada de la función

$$ f(x) = \sin(x) e^{-\frac{x}{4}} $$

usando el método de diferencias centradas en el intervalo \([0, 10]\), con 500 puntos.

### Desarrollo

La función que vamos a derivar es:

$$ f(x) = \sin(x) e^{-\frac{x}{4}} $$

Su derivada analítica es:

$$ f'(x) = \cos(x) e^{-\frac{x}{4}} - \frac{1}{4} \sin(x) e^{-\frac{x}{4}} $$




```python
# Paso 1: Importación de librerías

import numpy as np
import matplotlib.pyplot as plt
```

```python
# Definir la función f(x)
def f(x):
    return np.sin(x) * np.exp(-x / 4)

# Derivada analítica de f(x)
def f_prime_analitica(x):
    return np.cos(x) * np.exp(-x / 4) - (1 / 4) * np.sin(x) * np.exp(-x / 4)
```

```python
# Aproximación numérica de la derivada por diferencias centradas
def derivada_numerica(x, f, h=1e-5):
    return (f(x + h) - f(x - h)) / (2 * h)
```

```python
# Crear un conjunto de puntos
x = np.linspace(0, 10, 500)

# Calcular la derivada numérica
derivada_numerica_vals = derivada_numerica(x, f)

# Calcular la derivada analítica
derivada_analitica_vals = f_prime_analitica(x)
```

```python
# Graficar resultados
plt.figure(figsize=(10, 6))
plt.plot(x, derivada_numerica_vals, label="Derivada Numérica", linestyle='--')
plt.plot(x, derivada_analitica_vals, label="Derivada Analítica", linestyle='-')
plt.xlabel("x")
plt.ylabel("f'(x)")
plt.title("Comparación de Derivadas Numérica y Analítica")
plt.legend()
plt.grid(True)
plt.show()
```

## Ejercicio 5: Integración Numérica - Trapecio y Simpson

### Objetivo

Implementar la regla del trapecio y la regla de Simpson para realizar la integración numérica de la función 

$$ f(x) = \sin(x) e^{-\frac{x}{4}} $$

en el intervalo \([0, 10]\).

### Desarrollo



```python
# Paso 1: Definir la función y el intervalo

# Definir la función f(x)
def f(x):
    return np.sin(x) * np.exp(-x / 4)

# Definir el intervalo [0, 10] y el número de puntos
x = np.linspace(0, 10, 500)
```

```python
# Regla del trapecio
def trapecio(f, x):
    h = x[1] - x[0]  # Suponiendo que los puntos son equidistantes
    return h * (np.sum(f(x)) - 0.5 * (f(x[0]) + f(x[-1])))
```

```python
# Regla de Simpson
def simpson(f, x):
    h = x[1] - x[0]
    return (h / 3) * (f(x[0]) + f(x[-1]) + 4 * np.sum(f(x[1:-1:2])) + 2 * np.sum(f(x[2:-2:2])))
```

```python
# Aplicar ambos métodos
integral_trapecio = trapecio(f, x)
integral_simpson = simpson(f, x)

print(f"Integral por Trapecio: {integral_trapecio}")
print(f"Integral por Simpson: {integral_simpson}")
```

## Ejercicio 7: Estimación de parámetros de una distribución normal

### Objetivo

Simular 10,000 muestras de una distribución normal \( N(\mu = 2.5, \sigma = 1.3) \), estimar los parámetros \( \mu \) y \( \sigma^2 \) utilizando fórmulas vectorizadas, y compararlos con `np.mean` y `np.var`.

### Desarrollo


```python
# Parámetros de la distribución normal
mu = 2.5
sigma = 1.3

# Simular 10,000 muestras
muestras = np.random.normal(mu, sigma, 10000)

# Estimar mu y sigma^2 utilizando fórmulas vectorizadas
mu_estimado = np.sum(muestras) / len(muestras)
sigma2_estimado = np.sum((muestras - mu_estimado)**2) / len(muestras)

# Comparar con las funciones de NumPy
mu_np = np.mean(muestras)
sigma2_np = np.var(muestras)

print(f"Estimación de mu: {mu_estimado}, np.mean: {mu_np}")
print(f"Estimación de sigma^2: {sigma2_estimado}, np.var: {sigma2_np}")
```
